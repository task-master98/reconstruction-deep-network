```python
rotation_key = '135'
if rotation_key in loaded_embeddings:
    embeddings = loaded_embeddings[rotation_key]
    print(f"Embeddings for rotation {rotation_key}:")
    print("Type of embeddings:", type(embeddings))
    print("Contents of embeddings:", embeddings)
    print("Embeddings shape:", embeddings.shape)
else:
    print(f"No embeddings found for rotation {rotation_key}")
```
